```python
import csv
import json
import numpy as np
import pandas as pd
import re
import utils
from collections import OrderedDict, Counter
import ast
import swifter
import spicy_words
```

```python
df_eu = pd.read_csv("filtered_eu.csv") # load data
df_eu["name_list"] = df_eu["name_list"].apply(ast.literal_eval)
```

```python

foo = spicy_words.financial

foo = [el.replace(" ", utils.word_separator) for el in foo]

spicy_re = re.compile("|".join(foo))

# df_spicy = df_eu[df_eu.swifter.apply(lambda x: re.search(spicy_re, x["name"]), axis=1)]


```

```python
spicy_dict = {word: {} for word in foo}

for index, row in df_eu.iterrows():
    for word in foo:
        segname = row["name"]
        provider = row["provider_name"]
        res = re.findall(word, segname)
        if res:
            res = set(res)
            for tag in res:
                if provider in spicy_dict[word]:
                    spicy_dict[word][provider].append(segname)
                else:
                    spicy_dict[word][provider] = [segname]
```
